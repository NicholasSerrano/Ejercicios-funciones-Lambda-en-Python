Escriba un programa en Python que genere una función lambda que permita multiplicar dos números.

# Ejercicio 2
Escriba una función de Python para dar la suma de los números de una lista.

````python
# función
def suma_lista(lista):
    suma = 0
    for numero in lista:
        suma += numero
    return suma

# Uso de la función
numeros = [1, 2, 3, 4, 5]
resultado = suma_lista(numeros)
print("La suma de los números en la lista es:", resultado)
````

# Ejercicio 3

Dada una lista y un número, cree una función que devuelva dos parámetros:

1) Si está dicho número en esa lista. En dicho caso Devolverá `TRUE`. En caso contrario, `False`.

2) Si la longitud de la lista coincide con dicho número. En dicho caso devolverá `TRUE`. En caso contrario, `False`.

````python
# función
def comprobar_numero_y_longitud(lista, numero):
    esta_en_lista = numero in lista
    coincide_longitud = len(lista) == numero
    return esta_en_lista, coincide_longitud

# Uso de la función
mi_lista = [1, 2, 3, 4, 5]
mi_numero = 3
resultado = comprobar_numero_y_longitud(mi_lista, mi_numero)
print("Respuestas:", resultado)
````

# Ejercicio 4

Dada una lista, cree dos funciones, una que devuelva el mayor número y otra que devuelva en menor número.



````python
# función A
def encontrar_mayor(lista):
    if not lista:  # Si la lista está vacía
        return None
    mayor = lista[0]
    for numero in lista:
        if numero > mayor:
            mayor = numero
    return mayor

# función B
def encontrar_menor(lista):
    if not lista:  # Si la lista está vacía
        return None
    menor = lista[0]
    for numero in lista:
        if numero < menor:
            menor = numero
    return menor

# Uso de la funciónes A y B
mi_lista = [10, 25, 8, 15, 1]
mayor_numero = encontrar_mayor(mi_lista)
menor_numero = encontrar_menor(mi_lista)
print("El mayor número en la lista es:", mayor_numero)
print("El menor número en la lista es:", menor_numero)

````

#Ejercicio 7
Cree una función para contar la cantidad de números negativos de una lista.


````python
def contar_negativos(lista):
    contador = 0
    for numero in lista:
        if numero < 0:
            contador += 1
    return contador

# Ejemplo de uso
mi_lista = [10, -5, 8, -15, 3, -2, 0]
cantidad_negativos = contar_negativos(mi_lista)
print("La cantidad de números negativos en la lista es:", cantidad_negativos)

````

# Ejercicio 8

Cree una función para contar la cantidad de números impares de una lista.


````python
# FUNCIÓN
def contar_impares(lista):
    contador = 0
    for numero in lista:
        if numero % 2 != 0:
            contador += 1
    return contador

# uso de  la función
mi_lista = [10, 5, 8, 15, 3, 2, 0]
cantidad_impares = contar_impares(mi_lista)
print("La cantidad de números impares en la lista es:", cantidad_impares)
````

#Ejercicio 9

Dada una palabra, cuente el número de vocales que tiene.



````python
def contar_vocales(palabra):
    # Convertir la palabra a minúsculas para hacer la comparación más sencilla
    palabra = palabra.lower()
    # Definir una lista de vocales
    vocales = ['a', 'e', 'i', 'o', 'u']
    # Inicializar un contador para las vocales
    contador_vocales = 0
    # Iterar sobre cada letra de la palabra
    for letra in palabra:
        # Verificar si la letra es una vocal
        if letra in vocales:
            contador_vocales += 1
    return contador_vocales

# Palabra a analizar
palabra = "mercadona"

# Contar el número de vocales en la palabra
numero_vocales = contar_vocales(palabra)

print("El número de vocales en la palabra '{}' es: {}".format(palabra, numero_vocales))


````

````python
import array

cadena = "abcdefghijklmnopqrstuvwxyz"

# Convertir la cadena a un array de caracteres
arr = array.array('u', cadena)

# Revertir el array
arr.reverse()

# Convertir el array revertido de nuevo a una cadena
cadena_revertida = ''.join(arr)

print("Cadena original:", cadena)
print("Cadena revertida:", cadena_revertida)
````

# Ejercicio de ampliación 1
Modifique el programa de la contraseña para dar al usuario solo 3 intentos. Si pasados dichos intentos el usuario no ha escrito bien la contraseña el programa acabará.

````python
# Definir la contraseña correcta
contrasena_correcta = "123"

# Solicitar el nombre de usuario
nombre_usuario = input("Hola, ¿cuál es tu nombre de usuario? ")

# Inicializar el contador de intentos
intentos = 0

# Permitir un máximo de 3 intentos para la contraseña
while intentos < 3:
    # Solicitar la contraseña
    contrasena = input("Hola, {}! Por favor, introduce tu contraseña: ".format(nombre_usuario))

    # Verificar si la contraseña es correcta
    if contrasena == contrasena_correcta:
        print("¡Bienvenido, {}!".format(nombre_usuario))
        break
    else:
        intentos += 1
        intentos_restantes = 3 - intentos
        print("Contraseña incorrecta. Te quedan {} intentos.".format(intentos_restantes))

# Si el usuario no ha ingresado la contraseña correcta después de 3 intentos, mostrar un mensaje y terminar el programa.
if intentos == 3:
    print("Has agotado tus intentos. El programa ha terminado.")
````

````python
# Definir la contraseña
contrasena = "secreto"

# Inicializar el contador de intentos
intentos = 0

# Permitir un máximo de 3 intentos
while intentos < 3:
    # Pedir al usuario que ingrese la contraseña
    entrada = input("Ingrese la contraseña: ")

    # Verificar si la contraseña es correcta
    if entrada == contrasena:
        print("¡Contraseña correcta! Bienvenido.")
        break
    else:
        intentos += 1
        intentos_restantes = 3 - intentos
        print("Contraseña incorrecta. Te quedan {} intentos.".format(intentos_restantes))

# Si el usuario no ha ingresado la contraseña correcta después de 3 intentos, mostrar un mensaje y terminar el programa.
if intentos == 3:
    print("Has agotado tus intentos. El programa ha terminado.")

````
